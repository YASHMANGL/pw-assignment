```python
# Q.1 Can you explain the logic and working of the Tower of Hanoi algorithm by writing a Java program? 
# How does the recursion work, and how are the movements of disks between rods accomplished?

def tower_of_hanoi(n, source, auxiliary, destination):
    if n == 1:
        print(f"Move disk 1 from {source} to {destination}")
        return
    tower_of_hanoi(n-1, source, destination, auxiliary)
    print(f"Move disk {n} from {source} to {destination}")
    tower_of_hanoi(n-1, auxiliary, source, destination)

# Example: Move 3 disks from A to C
tower_of_hanoi(3, 'A', 'B', 'C')

```

```python
# Q.2 Given two strings word1 and word2, return the minimum number of operations required to convert word1  to word2.
# Example 1:
# Input: word1 = "horse", word2 = "ros"
# Output: 3
# Explanation: 
# horse -> rorse (replace 'h' with 'r')
# rorse -> rose (remove 'r')
# rose -> ros (remove 'e')
# Example 2:
# Input: word1 = "intention", word2 = "execution"
# Output: 5
# Explanation: 
# intention -> inention (remove 't')
# inention -> enention (replace 'i' with 'e')
# enention -> exention (replace 'n' with 'x')
# exention -> exection (replace 'n' with 'c')
# exection -> execution (insert 'u')

def min_distance(word1, word2):
    # Create memoization table as a dict
    memo = {}

    def dp(i, j):
        if (i, j) in memo:
            return memo[(i, j)]
        # If word1 is empty, insert all of word2
        if i == len(word1):
            return len(word2) - j
        # If word2 is empty, remove all of word1
        if j == len(word2):
            return len(word1) - i
        if word1[i] == word2[j]:
            ans = dp(i+1, j+1)
        else:
            insert = 1 + dp(i, j+1)
            delete = 1 + dp(i+1, j)
            replace = 1 + dp(i+1, j+1)
            ans = min(insert, delete, replace)
        memo[(i, j)] = ans
        return ans

    return dp(0, 0)

# Example usage:
print(min_distance("horse", "ros"))  # Output: 3
print(min_distance("intention", "execution"))  
```

```python
# Q. 3 Print the max value of the array [ 13, 1, -3, 22, 5].
arr = [13, 1, -3, 22, 5]

max_val = arr[0]  # Assume first element is max
for num in arr:
    if num > max_val:
        max_val = num

print("Max value:", max_val)  # Output: 22
```

```python
# Q.4 Find the sum of the values of the array [92, 23, 15, -20, 10].
arr = [92, 23, 15, -20, 10]

total_sum = 0
for num in arr:
    total_sum += num

print("Sum of array:", total_sum)  # Output: 120
```

```python
# Q.5 Given a number n. Print if it is an armstrong number or not.An armstrong number is a number if the sum
# of every digit in that number raised to the power of total digits in that number is equal to the number.
# Example : 153 = 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153 hence 153 is an armstrong number. (Easy)
# Input1 : 153
# Output1 : Yes

n = 153

# Count number of digits
temp = n
count = 0
while temp > 0:
    temp //= 10
    count += 1

# Calculate sum of digits^count
temp = n
sum_of_powers = 0
while temp > 0:
    digit = temp % 10
    sum_of_powers += digit ** count
    temp //= 10

# Check Armstrong condition
if sum_of_powers == n:
    print("Yes")  # Armstrong number
else:
    print("No")
```

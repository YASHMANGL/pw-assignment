```python
# 1.14.Consider the gravitational interactions between the Earth, Moon, and Sun in our solar system.
# Given:
# mass_earth = 5.972e24     # Mass of Earth in kilograms
# mass_moon = 7.34767309e22 # Mass of Moon in kilograms
# mass_sun = 1.989e30 # Mass of Sun in kilograms
# distance_earth_sun = 1.496e11 # Average distance between Earth and Sun in meters
# distance_moon_earth = 3.844e8 # Average distance between Moon and Earth in meters
# Tasks
#  Calculate the gravitational force between the Earth and the Sun
#  Calculate the gravitational force between the Moon and the Earth
#  Compare the calculated forces to determine which gravitational force is stronger
# Explain which celestial body (Earth or Moon) is more attracted to the other based on the comparison.

# Given data
mass_earth = 5.972e24  # in kg
mass_moon = 7.34767309e22  # in kg
mass_sun = 1.989e30  # in kg
distance_earth_sun = 1.496e11  # in meters
distance_moon_earth = 3.844e8  # in meters

# Gravitational constant
G = 6.67430e-11  # in N(m^2)/kg^2

# Calculate gravitational force: F = G * (m1 * m2) / r^2
force_earth_sun = G * (mass_earth * mass_sun) / (distance_earth_sun ** 2)
force_moon_earth = G * (mass_moon * mass_earth) / (distance_moon_earth ** 2)

print(f"Gravitational Force between Earth and Sun: {force_earth_sun:.2e} N")
print(f"Gravitational Force between Moon and Earth: {force_moon_earth:.2e} N")

# Comparison
if force_earth_sun > force_moon_earth:
    print("The gravitational force between the Earth and the Sun is stronger.")
else:
    print("The gravitational force between the Moon and the Earth is stronger.")

print("Both bodies attract each other with equal force, but the more massive body (Earth or Sun) influences the motion less.")
```

```python
# 4.Create a simple banking system using object-oriented concepts in Python. Design classes representing
# different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw,
# and balance inquiry. Utilize inheritance to manage different account types efficiently.
# Tasks
#  1. Define base class(es) for bank accounts with common attributes and methods
#  2. Implement subclasses for specific account types (e.g., SavingsAccount, CheckingAccount)
#  3. Provide methods for deposit, withdraw, and balance inquiry in each subclass
#  4. Test the banking system by creating instances of different account types and performing transactions.

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}. New balance = ${self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance = ${self.balance}")
        else:
            print("Insufficient funds.")

    def get_balance(self):
        return self.balance


class SavingsAccount(BankAccount):
    def __init__(self, owner, balance=0, interest_rate=0.02):
        super().__init__(owner, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest added: ${interest:.2f}. New balance = ${self.balance:.2f}")


class CheckingAccount(BankAccount):
    def __init__(self, owner, balance=0, overdraft_limit=500):
        super().__init__(owner, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance = ${self.balance}")
        else:
            print("Overdraft limit exceeded.")


# Testing the banking system
savings = SavingsAccount("Alice", 1000)
checking = CheckingAccount("Bob", 500)

savings.deposit(500)
savings.add_interest()

checking.withdraw(700)
checking.withdraw(400)
```

```python
# 7.Write a code for Hotel Management System using OOPS 
#  Create a Room class that has attributes such as room number, room type, rate, and availability (private)
#  Implement methods to book a room, check in a guest, and check out a guest
#  Use encapsulation to hide the room's unique identification number
# Inherit from the Room class to create a SuiteRoom class and a StandardRoom class, each with their own
# specific attributes and methods.

class Room:
    __room_id_counter = 1

    def __init__(self, room_number, room_type, rate, availability=True):
        self.__room_id = Room.__room_id_counter
        Room.__room_id_counter += 1
        self.room_number = room_number
        self.room_type = room_type
        self.rate = rate
        self.__availability = availability

    def book_room(self):
        if self.__availability:
            self.__availability = False
            print(f"Room {self.room_number} is booked.")
        else:
            print(f"Room {self.room_number} is not available.")

    def check_in(self):
        if not self.__availability:
            print(f"Checked into room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is not booked yet.")

    def check_out(self):
        if not self.__availability:
            self.__availability = True
            print(f"Checked out of room {self.room_number}. Room is now available.")
        else:
            print(f"Room {self.room_number} was not occupied.")

    def is_available(self):
        return self.__availability

class SuiteRoom(Room):
    def __init__(self, room_number, rate, has_jacuzzi=True):
        super().__init__(room_number, room_type="Suite", rate=rate)
        self.has_jacuzzi = has_jacuzzi

class StandardRoom(Room):
    def __init__(self, room_number, rate):
        super().__init__(room_number, room_type="Standard", rate=rate)


# Testing the Hotel system
suite = SuiteRoom(101, 250)
standard = StandardRoom(102, 100)

suite.book_room()
suite.check_in()
suite.check_out()
standard.check_in()
```

```python
# 8.Write a code for Fitness Club Management System using OOPS3
#  Create a Member class that has attributes such as name, age, membership type, and membership status
# (private)
#  Implement methods to register a new member, renew a membership, and cancel a membership
#  Use encapsulation to hide the member's unique identification number
#  Inherit from the Member class to create a FamilyMember class and an IndividualMember class, each with
# their own specific attributes and methods

class Member:
    __member_id_counter = 1

    def __init__(self, name, age, membership_type, status=True):
        self.__member_id = Member.__member_id_counter
        Member.__member_id_counter += 1
        self.name = name
        self.age = age
        self.membership_type = membership_type
        self.__status = status

    def register(self):
        self.__status = True
        print(f"{self.name} registered with {self.membership_type} membership.")

    def renew_membership(self):
        if self.__status:
            print(f"{self.name}'s membership renewed.")
        else:
            print(f"{self.name} is not an active member.")

    def cancel_membership(self):
        self.__status = False
        print(f"{self.name}'s membership is cancelled.")

    def is_active(self):
        return self.__status

class FamilyMember(Member):
    def __init__(self, name, age, membership_type, family_size):
        super().__init__(name, age, membership_type)
        self.family_size = family_size

class IndividualMember(Member):
    def __init__(self, name, age, membership_type):
        super().__init__(name, age, membership_type)


# Testing Fitness Club
member1 = IndividualMember("Alice", 30, "Gold")
member1.register()
member1.renew_membership()
member1.cancel_membership()
member1.renew_membership()
```

```python
# 24. Calculate coefficient of correlation between the marks obtained by 10 students in Accountancy andstatistics:
# Use Karl Pearson’s Coefficient of Correlation Method to find it.

import numpy as np

# Data
accountancy = [45, 70, 65, 30, 90, 40, 50, 75, 85, 60]
statistics  = [35, 90, 70, 40, 95, 40, 60, 80, 80, 50]

# Calculate means
mean_x = np.mean(accountancy)
mean_y = np.mean(statistics)

# Calculate deviations
dev_x = [x - mean_x for x in accountancy]
dev_y = [y - mean_y for y in statistics]

# Numerator: sum of products of deviations
numerator = sum(dx * dy for dx, dy in zip(dev_x, dev_y))

# Denominator: product of standard deviations
denominator = (sum(dx**2 for dx in dev_x) * sum(dy**2 for dy in dev_y)) ** 0.5

# Karl Pearson's correlation coefficient
r = numerator / denominator

print(f"Pearson's correlation coefficient: {r:.3f}")
```

```python
# 31. The mean of a distribution is 60 with a standard deviation of 10. Assuming that the distribution is normal, what percentage of items be (i) between 60 and 72, (ii) between 50 and 60, (iii) beyond 72 and (iv) between 70 and 80?
from scipy.stats import norm

mu = 60
sigma = 10

# i) between 60 and 72
p_60_72 = norm.cdf(72, mu, sigma) - norm.cdf(60, mu, sigma)
print("Between 60 and 72:", round(p_60_72*100, 2), "%")

# ii) between 50 and 60
p_50_60 = norm.cdf(60, mu, sigma) - norm.cdf(50, mu, sigma)
print("Between 50 and 60:", round(p_50_60*100, 2), "%")

# iii) beyond 72 (greater than)
p_beyond_72 = 1 - norm.cdf(72, mu, sigma)
print("Beyond 72:", round(p_beyond_72*100, 2), "%")

# iv) between 70 and 80
p_70_80 = norm.cdf(80, mu, sigma) - norm.cdf(70, mu, sigma)
print("Between 70 and 80:", round(p_70_80*100, 2), "%")
```

```python
# 32. 15000 students sat for an examination. The mean marks was 49 and the distribution of marks had a standard deviation of 6. Assuming that the marks were normally distributed what proportion of students scored (a) more than 55 marks, (b) more than 70 marks
from scipy.stats import norm

mu = 49
sigma = 6
n = 15000

# a) More than 55 marks
z_55 = (55 - mu) / sigma
proportion_55 = 1 - norm.cdf(55, mu, sigma)
print("More than 55 marks:", round(proportion_55 * n))

# b) More than 70 marks
z_70 = (70 - mu) / sigma
proportion_70 = 1 - norm.cdf(70, mu, sigma)
print("More than 70 marks:", round(proportion_70 * n))
```

```python
# 34. What is the statistical hypothesis? Explain the errors in hypothesis testing.b)Explain the Sample. What are Large Samples & Small Samples?
from scipy.stats import chi2

n = 25
sample_std = 9.0
pop_std = 10.5
alpha = 0.05

chi_sq_stat = (n-1)*(sample_std**2) / (pop_std**2)
chi_sq_critical = chi2.ppf(1-alpha, n-1)

print("Chi-square statistic:", round(chi_sq_stat,2))
print("Critical value (alpha=0.05):", round(chi_sq_critical,2))
if chi_sq_stat < chi_sq_critical:
    print("Accept null hypothesis: population std is 10.5")
else:
    print("Reject null hypothesis")
```
